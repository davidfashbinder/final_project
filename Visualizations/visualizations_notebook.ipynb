```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Files to load. Here is where we will add our two datasets from database
salaries_to_load = "Resources/Salaries.csv"
people_to_load = "Resources/People.csv"
batting_to_load = "Resources/Batting.csv"
```

```python
salaries_df = pd.read_csv(salaries_to_load)
salaries_df.head(10)
```

```python
people_df = pd.read_csv(people_to_load)
people_df.head(10)
```

```python
batting_df = pd.read_csv(batting_to_load)
batting_df.head(10)
```

```python
# Combine the data into a single dataset
pd.set_option('display.max_columns', None)
baseball_df = pd.merge(salaries_df, batting_df, how="left", on=["playerID", "playerID"])
baseball_df
```

```python
# Complete year range 
baseball_df["yearID_x"].unique()
```

```python
# Set the x-axis to the year
x_axis = baseball_df['yearID_x']

# Set the y-axis to the salary
y_axis = baseball_df['salary']
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
avg_salaries_by_year = baseball_df.groupby(["yearID_x"]).mean()["salary"]

avg_salaries_by_year_formatted = avg_salaries_by_year.round(decimals=0)

avg_salaries_by_year_formatted
```

```python
# Set the x-axis to the year
x_axis = baseball_df['yearID_x'].unique()

# Set the y-axis to the salary
y_axis = avg_salaries_by_year_formatted
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="green", linewidth=2)
# Create labels for the x and y axes.
plt.xlabel("Year")
plt.ylabel("Salaries (Millions)")
# Create a title.
plt.title("Avg Salaries by Year")
# Add a grid.
plt.grid()

#Saving figure
plt.savefig("Images/avg_salaries.png")
```

# Showcases the general increase in baseball player salaries over time

```python
avg_ab_by_player = baseball_df.groupby(["playerID"]).mean()["AB"]

avg_ab_by_player
```

```python
# Create the Urban city DataFrame.
baseball_2016_df = baseball_df[baseball_df["yearID_x"] == 2016]
baseball_2016_df.head()
```

```python
# team at bats average
ab_mean_2016 = baseball_2016_df.groupby(["teamID_x"]).mean()["AB"]
ab_mean_2016.head()
```

```python
# team average salary
salary_mean_2016 = baseball_2016_df.groupby(["teamID_x"]).mean()["salary"]
salary_mean_2016 = salary_mean_2016.round(decimals=0)
salary_mean_2016.head()
```

```python
# team runs average
runs_mean_2016 = baseball_2016_df.groupby(["teamID_x"]).mean()["R"]
runs_mean_2016.head()
```

```python
# Build the scatter plots for urban cities.
plt.scatter(ab_mean_2016,
      salary_mean_2016,
      s=10*runs_mean_2016, c="blue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Runs")
plt.title("Teams At-Bats vs Salary")
plt.ylabel("Salaries (Millions)")
plt.xlabel("Average At-Bats")
plt.grid(True)
# Add the legend.
plt.legend()

#Saving figure
plt.savefig("Images/team_scatter.png")
```

# Broad view of At-bats, Runs and Salaries by teams. Salaries appear less affected by how susbtantial the At-bats and Runs are

```python
# player at bats average
ab_mean_2016 = baseball_2016_df.groupby(["playerID"]).mean()["AB"]
ab_mean_2016.head()
```

```python
# player average salary
salary_mean_2016 = baseball_2016_df.groupby(["playerID"]).mean()["salary"]
salary_mean_2016 = salary_mean_2016.round(decimals=0)
salary_mean_2016.head()
```

```python
# player runs average
runs_mean_2016 = baseball_2016_df.groupby(["playerID"]).mean()["R"]
runs_mean_2016.head()
```

```python
# Build the scatter plot
plt.scatter(ab_mean_2016,
      salary_mean_2016,
      s=runs_mean_2016, c="green",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Runs")
plt.title("Players At-Bats vs Salary")
plt.ylabel("Salaries (Millions)")
plt.xlabel("Average At-Bats")
plt.grid(True)
# Add the legend.
plt.legend()

#Saving figure
plt.savefig("Images/player_scatter.png")
```

# Above image shows a correlation between higher at-bats scoring more runs however we still have players that make equal or higher salary to those with higher at-bat to runs ratio

```python
# Set the x-axis to a list of strings for each feature.
x_axis = ['yearID', 'birthYear', 'teamID_x', 'SO', 'AB', 'G', 'H', 'R', 'RBI', 'HR', 'awardID', 'lgID']
# Set the y-axis to a list of weighted importances
y_axis = [0.50, 0.19, 0.09, 0.03, 0.03, 0.03, 0.03, 0.03, 0.02, 0.02, 0.02, 0.01]
```
